```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('gorjetas.csv')
```

#### a) criar um gr√°fico de densidade do total de gorjetas por sexo;

```python
plt.figure(figsize = (10,6))
sns.jointplot(x = 'sexo', y = 'gorjeta', data = df)

plt.xlabel('Sexo')
plt.ylabel('Gorjeta')
plt.show()
```

#### b) criar um gr√°fico de densidade do total de gorjetas pela quantidade de pessoas na mesa

```python
plt.figure(figsize = (10,6))
sns.kdeplot(x = 'quantidade', y = 'gorjeta', data = df)

plt.xlabel('quantidade')
plt.ylabel('gorjeta')
plt.show()
```

#### c) criar um gr√°fico de densidade do total de gorjetas por dia da semana

```python
plt.figure(figsize = (10,6))
sns.jointplot(x = 'dia', y = 'gorjeta', data = df)

plt.xlabel('Dias')
plt.ylabel('gorjeta')
plt.show()
```
